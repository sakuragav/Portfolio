```python
import pandas as pd
import seaborn as sns
import numpy as np
# Evaluate what appropriate visualization would be for a global pattern to depict the sepal length
# and sepal width features. Make your to include the code and output.
df = pd.read_csv('https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/0e7a9b0a5d22642a06d3d5b9bcbad9890c8ee534/iris.csv')
sns.kdeplot(df.sepal_length, df.sepal_width, shade = True)
```

```python
df['sepallength'] = np.floor(df.sepal_length)
df['sepallength'] = df['sepallength'].astype(int)
df.tail()
sns.boxplot(x = 'sepallength', y = 'sepal_width', data = df, hue = 'species')
```

```python

sns.catplot(data = df, x = "sepallength", y = "sepal_width", hue = "species", kind = "violin", aspect = 1.5)
```
